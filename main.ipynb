```python
def Overview(data) :
    profile = ProfileReport(data)
    profile.to_file("Analysis.html")

train_data = pd.read_csv('data/train.csv', low_memory=False)
store_data = pd.read_csv('data/store.csv')
```

```python
# Changing the data of columns

# Converting in datetime format

data['Date'] = pd.to_datetime(data['Date'])
```

```python
max(data.Store)

# So Convert it into int32

data['Store'] = data.Store.astype('int32')
```

```python
def ConvertDataType(data) : 
    for column in data.columns :
        if data[column].dtype == 'int64' :
            if max(data[column]) <= 8000 :
                data[column] = data[column].astype('int32')
        elif data[column].dtype == 'float64' :
            if max(data[column]) <= 8000 :
                data[column] = data[column].astype('float32')
    return data
```

```python
data = ConvertDataType(data)
```

```python
data.info()
```

```python
# Time Series Data analysis
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize': (8, 6)})
# sns.lineplot(x='Date', y='Sales', data=data)
```

```python
data['Date'].dt.year.unique()
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

monthly_average_sales = data.groupby(['Year', 'Month'])['Sales'].mean().reset_index()
```

```python
monthly_average_sales
```

```python
sns.set(rc={'figure.figsize': (10, 8)})

sns.lineplot(data=monthly_average_sales, x='Year', y='Sales', label='Average')

plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
daily_data = data.resample('D', on='Date').sum()['Sales']
```

```python
monthly_data_mean = daily_data.rolling(window=30).mean()
```

```python
weekly_data_mean = daily_data.rolling(window=7).mean()
```

```python
sns.lineplot(data=daily_data, x=daily_data.index, y=daily_data.values, label = "Original Data")
sns.lineplot(data=weekly_data_mean, x=weekly_data_mean.index, y= weekly_data_mean.values, label="Average Weekly Values")
sns.lineplot(data=monthly_data_mean, x=monthly_data_mean.index, y=monthly_data_mean.values, label = "Average Monthly Data")
```

```python
# Looking the sales of every month 

yearly_data = data.resample('M', on='Date').sum()['Sales'].reset_index()
yearly_data['year'] = yearly_data['Date'].dt.year
for year in list(yearly_data['year'].unique()) :
    year_on = yearly_data[yearly_data['Date'].dt.year == year]
    sns.lineplot(data=year_on, x=year_on['Date'].dt.month, y='Sales', label=str(year))
```
